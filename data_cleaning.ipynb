```python
df_pre = pd.read_csv("goodreads_dataset.csv", encoding='ISO-8859-1')
df_pre.shape
```

```python
df_tags = pd.read_csv("tags.csv", encoding='ISO-8859-1')
df_tags.shape
```

```python
df_tags=df_tags.dropna()
df_tags.head
```

```python
df_book_tags = pd.read_csv("book_tags.csv", encoding='ISO-8859-1')
df_book_tags.shape
```

```python
df_book_tags['tag_id'] = df_book_tags['tag_id'].map(df_tags.set_index('tag_id')['tag_name'])
df_book_tags.head
```

```python
from collections import defaultdict
d = defaultdict(list)

for i, tag in zip(df_book_tags['goodreads_book_id'], df_book_tags['tag_id']):
    d[i].append(tag)
```

```python
genre = ['fantasy', 'young-adult', 'romance', 'mystery/action/thriller/adventure', 'comedy', 'crime', 'economics', 'politics', 'biography', 'sci-fi', 'horror', 'history', 'spiritual', 'children', 'comics', 'science', 'society', 'fiction', 'non-fiction']
```

```python
def remove_element(a, e):
    try:
        a.remove(e)
    except ValueError:
        pass
```

```python
for i, tags in d.items():
    if i == 30253700:
        print(i, tags)
```

```python
tag_dict={}
for i, tags in d.items():
#     print(i, tags)
    genre_list = list(genre)
    c=0
    final_genres=[]
    for j in tags:
        for k in genre_list:
            if k in str(j):
                if not 'fiction' in j:
                    remove_element(genre_list, k)
                else:
                    if 'non-fiction' in j:
                        remove_element(genre_list, 'non-fiction')
                        j=j.replace('non-fiction', '')
                        k='non-fiction'
                    else:
                        remove_element(genre_list, 'fiction')
                        j=j.replace('fiction', '')
                        k='fiction'
#                 print("genre: ",genre_list)
                if k not in final_genres:
                    final_genres.append(k)
                    c=c+1
#                 print(c,k,j, final_genres)     
        if c==3:
            break
    if len(final_genres):
        tag_dict[i]=final_genres
    
#     if i > 20000:
#         break
tag_dict
    
```

```python
df_genre_mapped= pd.DataFrame(tag_dict.items())
df_genre_mapped.index.name='id genre'
df_genre_mapped.columns = ['id', 'genre']
df_genre_mapped.head
```

```python
df = pd.merge(df_pre, df_genre_mapped, on='id')
display(df)
```
